```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

Load the dataset

```python
X = np.loadtxt('gatlin.csv', delimiter=',')
print('X is of shape {}'.format(X.shape))
print(X)
```

## TODO: Perform SVD on the X matrix
Instructions: Perform SVD decomposition of matrix X. 
## Save the three factors in variables U, S and V

```python
# YOUR CODE HERE

U, S, V = np.linalg.svd(X)
```

```python
# Plot the original image
plt.figure(1)
plt.imshow(X, cmap=cm.Greys_r)
plt.title('Original image (rank 480)')
plt.axis('off')
plt.draw()
```

Matrix reconstruction using the top k = [10, 20, 50, 100, 200] singular values

## TODO: Create four matrices X10, X20, X50, X100, X200 for each low rank approximation

```python
# YOUR CODE HERE
vec_10 = S[:10]
vec_20 = S[:20]
vec_50 = S[:50]
vec_100 = S[:100]
vec_200 = S[:200]

diag_10 = np.diag(vec_10)
diag_20 = np.diag(vec_20)
diag_50 = np.diag(vec_50)
diag_100 = np.diag(vec_100)
diag_200 = np.diag(vec_200)

X10 = np.dot(np.dot(U[:, :10], diag_10), V[:10, :])
X20 = np.dot(np.dot(U[:, :20], diag_20), V[:20, :])
X50 = np.dot(np.dot(U[:, :50], diag_50), V[:50, :])
X100 = np.dot(np.dot(U[:, :100], diag_100), V[:100, :])
X200 = np.dot(np.dot(U[:, :200], diag_200), V[:200, :])
```
